## These dims are created based on the values present in the Data Dictionary as per NYC Taxi

```python
s3_path = "s3://nyc-tlc-data-398563707364/silver/"
```

Payment Type Dim

```python
payment_data = [
    (0,"Flex Fare Trip"),(1, "Credit card"), (2, "Cash"), (3, "No charge"), 
    (4, "Dispute"), (5, "Unknown"), (6, "Voided trip")
]
spark.createDataFrame(payment_data, ["payment_type_id", "payment_type_name"]) \
     .write.format("delta").mode("overwrite").option("path", s3_path+"dim_payment_types/").saveAsTable("nyc_taxi.silver.dim_payment_types")
```

Rate Code Dim

```python

rate_code_data = [
    (1, "Standard rate"), (2, "JFK"), (3, "Newark"), 
    (4, "Nassau or Westchester"), (5, "Negotiated fare"), (6, "Group ride"),(99, "Null/unknown")
]
spark.createDataFrame(rate_code_data, ["rate_code_id", "rate_code_name"]) \
     .write.format("delta").mode("overwrite").option("path", s3_path+"dim_rate_codes/").saveAsTable("nyc_taxi.silver.dim_rate_codes")
```

Vendor Dim

```python
vendor_data = [
    (1, "Creative Mobile Technologies, LLC"), (2, "Curb Mobility, LLC"),
     (6, "Myle Technologies Inc"),(99, "Helix")
]
spark.createDataFrame(vendor_data, ["vendor_id", "vendor_name"]) \
     .write.format("delta").mode("overwrite").option("path", s3_path+"dim_vendors/").saveAsTable("nyc_taxi.silver.dim_vendors")
```

Trip Type Dim

```python
trip_type_data = [
    (1, "Street-hail"), (2, "Dispatch")
]
spark.createDataFrame(trip_type_data, ["trip_type_id", "trip_type_name"]) \
     .write.format("delta").mode("overwrite").option("path", s3_path+"dim_trip_types/").saveAsTable("nyc_taxi.silver.dim_trip_types")
```
